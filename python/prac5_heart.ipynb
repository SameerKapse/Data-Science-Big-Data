```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('heart.csv')
df
```

```python
df.columns
```

```python
df.describe()
```

```python
#Data cleaning heart
df.dropna(thresh=0.5 * len(df),axis=1)
```

```python
df1=df[['age','sex','cp','ca']].loc[0:15]
df1
```

```python
df2=df[['age','sex','cp','ca']].loc[16:30]
df2
```

```python
merge=pd.merge(df1,df2,on='age',how='inner')
merge
```

```python
#Data Transformation
df['target']=df['target'].apply(lambda x:1 if x>0 else 0)
df
```

```python
#Error Correction
df = df.applymap(lambda x: df.mean() if x < 0 else x)
df
```

```python
#Model Building
from sklearn.model_selection import train_test_split
X = merge.drop(['age'], axis=1)
y = merge['age']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Build the logistic regression model
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# Evaluate the model
from sklearn.metrics import classification_report, confusion_matrix
y_pred = logreg.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
