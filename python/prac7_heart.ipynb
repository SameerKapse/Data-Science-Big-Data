```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('heart.csv')
df
```

```python
#BoxPlot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['age'])
plt.xlabel('age')
plt.title('Boxplot of Age')
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['fbs'])
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('Histogram of age')
plt.show()
```

```python
#Single Line Graph
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['age'], marker='o')
plt.xlabel('Index')
plt.ylabel('age')
plt.title('age over time')
plt.show()
```

```python
df1=df[['slope']]
df1
```

```python
#MultiLine Graph
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['age'], marker='o', label='age')
plt.plot(df1.index, df1['slope'], marker='o', label='slope')
plt.xlabel('Index')
plt.ylabel('Metrics')
plt.title('Multiple Metrics over time')
plt.legend()
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='trestbps', hue='slope')
plt.xlabel('age')
plt.ylabel('trestbps')
plt.title('Scatter Plot of trestbpsh vs. age')
plt.show()
```

```python
#Bar Graph
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='age', y='trestbps', hue='slope')
plt.xlabel('age')
plt.ylabel('trestbps')
plt.title('Bar Graph of trestbps and age')
plt.show()
```

```python
#Box Plot with Notch
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='slope', y='age', notch=True)
plt.xlabel('slope')
plt.ylabel('age')
plt.title('Box Plot with Notch of age by slope')
plt.show()
```

```python
#Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr())
plt.title('Correlation Heatmap of Heat Metrics')
plt.show()
```

```python
#pie Chart
plt.figure(figsize=(8, 8))
category_counts = df['slope'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of slope')
plt.axis('equal')
plt.show()
```
